#Aqui irei ler um dataSet externo, renomearei seu cabeçalho, farei um split entre treino e teste e irei fazer as verificações de acurácia.

# 1 Importando dataSet

Usarei a biblioteca pandas para ler o arquivo


```
import pandas as pd

#Endereço do CSV (Dataset)
uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"

dataset = pd.read_csv(uri)

#Vou imprimir o inicio do dataser so pra checar.

dataset.head()
```

#2 Renomeando o cabeçalho

Para tal, usarei a função rename() do pandas, que recebe um dicionario como parâmetro, para fazer a renomeação.

Link para saber mais sobre dicionário em python:
http://blog.alura.com.br/trabalhando-com-o-dicionario-no-python/

```
dicionario = {
     "home": "principal",
     "how_it_works": "como_funciona",
     "contact": "contato",
     "bought": "comprou"
}
#Essa funcao nao altera o dataset original. Mas retorna uma novo, com os mesmas instancias, com o cabeçalho renomeado.
#Entao tenho que salvar de novo em uma variavel. Vou aproveitar a mesma.

dataset = dataset.rename(columns = dicionario)

dataset.head()
```

#3 Fazendo o split do dataset em treino e teste

Será usado o train_test_split do  módulo model_select da biblioteca sklearn para fazer o split

Aqui temos alguns conceitos importantes. A funcao train_test_split embaralha as instancias do dataset antes de fazer os split entre treino e teste. Para podermos replicar o modelo, presciamos passar um SEED como parâmetro para a função. O SEED é como um oráculo que ordena os números aleátorios.

Depois, temos que garantir que treino e teste tenham a mesma proporção na diferenças entre a quantidade de classes. Por exemplo, se no treino temos 40 de uma classe e 20 de outra, seria interessante se a proporção da diferença entre a quantidade de classes no teste também fosse de 2:1.  Para isso ser alcançado, iremos passar o stratify = y, como parâmetro. Isso significa que a distribuição em relação a y, que são as classes, será proporcional entre treino e teste.

Para finalizar, o teste_size é a fatia do dataset que será destinada para teste o reste será para treino. É interessante a maioria ficar para treino. O teste_size recebe um número entre 0 e 1, que será porcentagem.

Passamos o x e y, além de todos os conceitos acima, como parâmetro pra funcão train_test_split.

```
from sklearn.model_selection import train_test_split

x = dataset[ ["principal", "como_funciona","contato"]] #Pegando todas instancias do dataset a partir dos cabecalhos
y = dataset["comprou"] #Pegando todas as classes do dataset

#A funcao train_test_split retorna quatro resultados nesta ordem: x de treino, x de teste, y de treino, y de teste.
#Entao vou salvar esse retorno em quatro variaveis

SEED = 0
fatia_teste = 0.25 #25% do dataset será para teste

x_treino, x_teste, y_treino, y_teste = train_test_split( x, y, random_state = SEED, test_size = fatia_teste, stratify = y)

print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(x_treino), len(x_teste)))



```

# 4 Criando o modelo

Será usado o LinearSVC do módulo svm do sklearn.

```
from sklearn.svm import LinearSVC

modelo = LinearSVC()

#Agora o modelo sera treinado com a funcao fit() que vai receber o x de treino e o y de treino como parametro

modelo.fit(x_treino, y_treino)
```

# 5 Fazendo as estimativas e verificando a acurácia

Para calcular a acurácia, será usado o accuracy_score do módulo metrics da biblioteca sklearn

```
from sklearn.metrics import accuracy_score

estimativas = modelo.predict(x_teste) # Aqui o modelo faz as estimativas das classes das instancias passados como parametro.

#Agora vou verificar a acurácia dessas estimativas, comparando estas com o gabarito usando a funcao accuracy_score().

acuracia = accuracy_score(y_teste, estimativas) 

#A funcao retorna um numero entre 0 e 1. Vou multiplicar por 100 para ficar num formato de porcentagm.

acuracia = acuracia * 100

print("A acurácia foi %.2f%%" % acuracia)
```
