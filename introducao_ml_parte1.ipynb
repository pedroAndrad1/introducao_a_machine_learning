O seguinte dataset será usado para esse treinamento.

```
#DataSet

# features (1 sim, 0 não)
# pelo longo? 
# perna curta?
# faz auau?
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

x_train = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
#1 para porco e 0 para cachorro
y_train = [1,1,1,0,0,0]
```

Criando o modelo (Estimador) usando o sklearn

```
#Criando o modelo (Estimador) usando o sklearn

#Primeiro, é preciso importar o modulo svm da biblioteca sklearn, la esta a classe LinearSVC que sera usada para aprender com o DataSet e estimar.

from sklearn.svm import LinearSVC

modelo = LinearSVC() #Instanciando o modelo

#Agora, o modelo precisa se ajustar ao DataSet usando a funcao fit() que recebe dois paramatros: os dados para treino, x_train, e as classes para treino, y_train.
#Seria o equivalente a exemplos e gabaritos.

modelo.fit(x_train, y_train)


#E retorna essas informções ai
```

Agora, vamos testar o modelo e ver a acurácia(accuracy #speakEnglish) do modelo criado e ajustado ao DataSet. Para isso, é preciso exemplos e gabaritos para testar o modelo. Os exemplos serão passados para o modelo e este fará estimativas, ou seja, vai dizer se cada instância é cachorro ou porco.

```
#Exempos e Gabaritos

misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

x_teste = [misterio1, misterio2, misterio3]
y_teste= [0, 1, 1]

#Agora, veremos oq o modelo responde com a funcao predict()

estimativas = modelo.predict(x_teste)

estimativas

#Lembrando que 0 pra cachorro e 1 para porco
```

Agora, veremos, calcularemos a taxa de acerto do modelo. Para tal, será usado o accuracy_score do modulo metrics da biblioteca sklearn

```
#Importando accuracy_score

from sklearn.metrics import accuracy_score

#O accuracy_score recebe dois paramatros, o gabarito, Y, e as estimativas feitas pelo modelo,
taxa_de_acerto = accuracy_score(y_teste, estimativas)

taxa_de_acerto
#O accuracy_score retorna um numero de 0 a 1 que representa a taxa de acerto
```
